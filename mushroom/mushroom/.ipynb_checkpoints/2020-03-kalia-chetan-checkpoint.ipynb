## Analysis 1
---
*Briefly explain analysis here*

checked the data type of the variables to ensure they are all catagorical
confirmed there are no missing/corrupt values
graphed the class variables to confirm a balanced ratio, otherwise have to use advanced analytics like minority sampling
checked for unique values to check if any catagories have one unique varibles, result : drop veil-type
since data is catagorical and target variable only has two classes, can use chi2 test
chi2 test gives significance of each variable, A high chi square value indicates that that variable is strongly associated with the target.
result: veil-type has no correlation since p value is zero, odor is very important,gill-color, ring-type, etc...

```python
### Code for Analysis 12 here ###
df = pd.read_csv('mushrooms.csv')
#split independant variables from dependant
x = df.loc[:, df.columns != "class"]
y = df["class"]
print(df.nunique().sort_values(ascending=False))
#since veil-type has one label, drop it
# check that all data is catagorical - then use chi squared test
df.info()
print(df.isna().sum())
alpha = 0.05
#chi2 table
c = list(df.columns)
c.remove('class')
pval=[]
chi=[]
for i in c:
    ct=pd.crosstab(df[i],columns=df['class'])
    chi2, p, ddof, expected = scipy.stats.chi2_contingency(ct)
    pval.append(p)
    chi.append(chi2)
chisq1=pd.DataFrame()
chisq1['Properties']=c
chisq1['P value']=pval
chisq1['Chi Square value']=chi
chisq1.sort_values('Chi Square value',ascending=False)
print(chisq1)

#drop veil-type
df = df.drop(["veil-type"], axis=1)
x = x.drop(["veil-type"], axis=1)
#make sure balanced class types
x = df['class']
ax = sns.countplot(x=x,data=df)



# all signifiant since p values less than 0.05, other than veil-type

```

This test is performed to translate from heat map to correlaiton matrix to get quantative data. Known as cramersv heatmap
score closer to 1 means strong association
drawback: since it is symmetrical, doesn't account for theils-u; doesn't show which factor impacts the other, just that they are both impacted strongly by eachother
improvement: apply uncertainity coefficient to the cramersv heat map
result: odor and class have very strong relationship
result: strong correlation between gill-attach & veil-color/stalk color
result: very weak correlation between gill-spacing and gill-attach
...

```python

# accounts for theils u - masks

def cramers_corrected_stat(x,y):
    
    confusion_matrix = pd.crosstab(x,y)
    chi2 = ss.chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2/n
    r, k = confusion_matrix.shape
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))    
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)
    
    return np.sqrt(phi2corr / min((kcorr-1), (rcorr-1)))

cols = list(df.columns.values)
corrM = np.zeros((len(cols),len(cols)))

# Calculate correlations of every combination of variables
for col1, col2 in itertools.combinations(cols, 2):

    A, B = df[col1], df[col2]
    
    idx1, idx2 = cols.index(col1), cols.index(col2)
    corrM[idx1, idx2] = cramers_corrected_stat(A,B)
    corrM[idx2, idx1] = corrM[idx1, idx2]

corr = pd.DataFrame(corrM, index=cols, columns=cols)

cmap = sns.cubehelix_palette(light=1, as_cmap=True)

# Draw the heatmap with the mask 
fig = plt.figure(figsize=(20, 20))

sns.heatmap(corr, cmap=cmap, square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True)

plt.show()

```
